4. Now in this we convert this into 2 dataframes, where i want to add the date and time and in other data frane i add username and content

**Done by using regex expression string or patterns**

```python
pattern=r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

**Under this sometimes in Python \d already assumes as digit in python so to avoid it we use r in begin which detects as raw string.**

5. Now in this we then split the messages using pattern designed and data initialized.

6. Now in this block output shows empty strings in the start.So to avoid it we use **slicing**.
7. Then extraction of dates using findall().

```python
messages = re.split(pattern,data)[1:]
messages
```

```python
dates = re.findall(pattern,data)
dates
```

8. Data Frame Display where 2 columns: user_message and date

```python
df=pd.DataFrame({'user_message':messages,'date':dates})
#Convert date column type
df['date']=pd.to_datetime(df['date'],format='%d/%m/%y, %H:%M - ')
df.head()
```

```python
print(f"The shape of the dataframe is: {df.shape}")
```

9. Seperate user name and message in seperate columns

```python
users=[]
messages=[]
for user_message_content in df['user_message']:
  parts = user_message_content.split(': ', 1)
  if len(parts) > 1:
    user_name = parts[0]
    message_content = parts[1]
    users.append(user_name)
    messages.append(message_content)
  else:
    users.append('group_notification')
    messages.append(user_message_content)
df['user']=users
df['message']=messages
df.drop(columns=['user_message'],inplace=True)
df.head()
```

7. 20 Most Common Words: Use Frequency of words present and find top 20

**Now here several stop words,media omitted ,etc. is found in it. So we have to remove it as it is n't need.**

**Remove group messages,stop words ,media omitted**

```python
temp=df[df['user']!='group_notification']
temp=temp[temp['message']!='<Media omitted>\n']
```

```python
f=open('/content/stop_hinglish.txt','r')
stopwords=f.read()
```

```python
words=[]
for message in temp['message']:
  for word in message.lower().split():
    if word not in stopwords:
      words.append(word)
```

```python
from collections import Counter
common=pd.DataFrame(Counter(words).most_common(20))
common
```

```python
plt.barh(common[0],common[1])
plt.show()
```

**HeatMap to detect users active in particular day and time**

```python
df
```

```python
period=[]
for hour in df[['day_name','hour']]['hour']:
  if hour==23:
    period.append(str(hour)+"-"+str('00'))
  elif hour==0:
    period.append(str('00')+"-"+str(hour+1))
  else:
    period.append(str(hour)+"-"+str(hour+1))
```

```python
df['period']=period
```

```python
df['period']=df['period'].astype(str)
```

```python
df.sample(5)
```

```python
import seaborn as sns
activity_heatmap=df.pivot_table(
    index='day_name',
    columns='period',
    values='message',
    aggfunc='count'
).fillna(0)
plt.figure(figsize=(20,6))
sns.heatmap(activity_heatmap)
plt.yticks(rotation='horizontal',fontsize=10)
plt.xticks(rotation='vertical',fontsize=10)
plt.show()
```
